```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# loading the dataset to a Pandas DataFrame, skipping lines with errors
credit_card_data = pd.read_csv('/content/creditcard.csv', on_bad_lines='skip')
# or to display a warning instead of skipping them, use:
# credit_card_data = pd.read_csv('/content/creditcard.csv', error_bad_lines=False)

# Continue with the rest of your code...
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
# Selecting only numerical features for calculating the mean
numerical_features = credit_card_data.select_dtypes(include=np.number).columns
credit_card_data.groupby('Class')[numerical_features].mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
# Convert relevant columns to numeric before calculating the mean
for column in new_dataset.columns:
    if new_dataset[column].dtype == 'object' and column != 'Class':  # Exclude the 'Class' column
        try:
            new_dataset[column] = pd.to_numeric(new_dataset[column])
        except ValueError:
            print(f"Warning: Column '{column}' could not be converted to numeric and will be excluded from the mean calculation.")

new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
